```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('iris.csv',names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
features = df[['petal_length','petal_width']]
features.head()
```

```python
%matplotlib inline
plt.scatter(features['petal_length'],features['petal_width'])
plt.xlabel('petal_length')
plt.ylabel('petal_width')
```

```python
kmc = KMeans(n_clusters = 3)
prediction = kmc.fit_predict(features)
prediction
```

```python
features['cluster']=prediction
features.head(10)
```

```python
features.cluster.unique()
```

```python
iris0 = features[features.cluster==0]
iris1 = features[features.cluster==1]
iris2 = features[features.cluster==2]
```

```python
iris0.head()
```

```python
iris1.head()
```

```python
iris2.head()
```

```python
kmc.cluster_centers_
```

```python
plt.scatter(iris0.petal_length,iris0.petal_width,color='red',label='iris0')
plt.scatter(iris1.petal_length,iris1.petal_width,color='green',label='iris1')
plt.scatter(iris2.petal_length,iris2.petal_width,color='purple',label='iris2')
plt.legend()
plt.xlabel('Petal_Length')
plt.ylabel('Petal_width')

```
